## 2.1 Criando um DataFrame Unificado

```python
# dados mais importantes -> PIB per capita
# Criando um merge do df anterior mais o outro dataframe que haviamos importado
DATA = pd.merge(df_pop,df_dtb,how='right',left_on='Cód.',right_on='Código Município Completo')
DATA
```

```python
# Agora fizemos mais um merge e criamos um grande e único dataframe DATA
DATA = pd.merge(DATA,df_mun,how='left', left_on = 'Município_x', right_on = 'Município')
DATA
```

```python
DATA = pd.merge(DATA,geo,how='outer',left_on='Cód.',right_on='GEOCODIGO_MUNICIPIO')
DATA
```

```python
# Visualizando as informações do df completo
DATA.info()
```

```python
# Tratando os valores NaN
municipios_sem_coordenadas = DATA[DATA[['LATITUDE', 'LONGITUDE']].isnull().any(axis=1)]

print(municipios_sem_coordenadas['Município_x'])
```

```python
# Lista dos municípios com NaN que vamos atualizar
municipios_para_atualizar = municipios_sem_coordenadas['Município_x'].tolist()

# Dicionário com as novas coordenadas
coordenadas = {
    "Mojuí dos Campos (PA)": {"LATITUDE": -2.17139, "LONGITUDE": -56.74500},
    "Balneário Rincão (SC)": {"LATITUDE": -28.8344, "LONGITUDE": -49.2361},
    "Pescaria Brava (SC)": {"LATITUDE": -28.4247, "LONGITUDE": -48.8956},
    "Pinto Bandeira (RS)": {"LATITUDE": -29.0978, "LONGITUDE": -51.4503},
    "Paraíso das Águas (MS)": {"LATITUDE": -19.0257, "LONGITUDE": -53.0102},
}

for municipio in municipios_para_atualizar:
    if municipio in coordenadas:
        DATA.loc[DATA['Município_x'] == municipio, 'LATITUDE'] = coordenadas[municipio]["LATITUDE"]
        DATA.loc[DATA['Município_x'] == municipio, 'LONGITUDE'] = coordenadas[municipio]["LONGITUDE"]
```

## 3.1 Análise do PIB

```python
plt.figure(figsize=(10,6))
DATA['PIB (Mil Reais)'].hist(bins=50, color='skyblue', edgecolor='black')

plt.title('Distribuição do PIB (Mil Reais)', fontsize=15)
plt.xlabel('PIB (Mil Reais)', fontsize=13)
plt.ylabel('Número de Municípios', fontsize=13)

plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
# Olhar valores PIB

import numpy as np
from scipy import stats

pib = DATA['PIB (Mil Reais)']

print("Média: ",np.mean(pib))
print(sum(pib)/len(pib))

print("Mediana: ",np.median(pib))
ss=sorted(pib)
print(ss[int(len(ss)/2)])

print("Moda: ",(stats.mode(pib,keepdims=True)[0])[0])
# print(stats.mode(s_int))
```

```python
print("Desvio padrão: ",np.std(pib))
pib_mu = pib-np.mean(pib)
print(np.sqrt(sum(pib_mu**2)/len(pib)))
print("Variância: ",np.var(pib))

print("Valor Máximo: ",np.max(pib))
print("Valor Minimo: ",np.min(pib))

print("Erro padrão: ",stats.sem(pib, ddof=0))
print(np.std(pib)/np.sqrt(len(pib)) )
```

```python
# Análise dos valores dos Quartis

# print(ss)

print("1o. Quartil:",np.quantile(pib,0.25))
# pos=int(len(ss)*0.25)
# print(pos,ss[pos])

print("2o. Quartil:",np.quantile(pib,0.5))
print("Mediana: ",np.median(pib))

print("3o. Quartil:",np.quantile(pib,0.75))
# print(ss[int(len(ss)*0.75)])
# pos=int(len(ss)*0.75)
# print(pos,ss[pos])

print("1o. Percetil:",np.quantile(pib,0.01))


# Distância Interquartil
print('Dist. inter quartil = ',np.quantile(pib,0.75)-np.quantile(pib,0.25));
```

```python
# Assimetria
DATA['PIB (Mil Reais)'].skew()

# A distribuição tem uma cauda direita mais longa.
# Indica que a maioria dos municípios tem um PIB baixo, mas há alguns municípios com PIB extremamente alto (outliers) que puxam a média para cima.
```

```python
# Curtose
DATA['PIB (Mil Reais)'].kurtosis()

#Este é um valor extremamente alto de curtose.
# Isso indica que a distribuição tem caudas muito pesadas e um pico muito pontiagudo.
# Em outras palavras, há uma presença significativa de outliers ou valores extremos na distribuição do PIB.
```

```python
# Buscando pelo comportamento dos dados do PIB, identificando outliers no processo.
q1 = np.quantile(pib,0.25)
q3 = np.quantile(pib,0.75)
d = q3-q1

pib[pib > q3+1.5*d]
```

```python
# Criar um dicionário mapeando estados para regiões
regioes = {
    'Norte': ['Rondônia', 'Acre', 'Amazonas', 'Roraima', 'Pará', 'Amapá', 'Tocantins'],
    'Nordeste': ['Maranhão', 'Piauí', 'Ceará', 'Rio Grande do Norte', 'Pernambuco', 'Paraíba', 'Sergipe', 'Alagoas', 'Bahia'],
    'Sudeste': ['Minas Gerais', 'Espírito Santo', 'Rio de Janeiro', 'São Paulo'],
    'Sul': ['Paraná', 'Santa Catarina', 'Rio Grande do Sul'],
    'Centro-Oeste': ['Mato Grosso do Sul', 'Mato Grosso', 'Goiás', 'Distrito Federal']
}

# Função para mapear estado para região
def estado_para_regiao(estado):
    for regiao, estados in regioes.items():
        if estado in estados:
            return regiao
    return None

# Adicionar a coluna de região
df['Região'] = df['Nome_UF'].apply(estado_para_regiao)
df_log['Região'] = df_log['Nome_UF'].apply(estado_para_regiao)
```

```python

# Calcular a média nacional do PIB total
media_nacional_pib_total = df['PIB (Mil Reais)'].sum() / df['Região'].nunique()

# Gráfico para PIB total por região
sns.barplot(x='Região', y='PIB (Mil Reais)', data=df, estimator=sum, ci=None)
plt.axhline(media_nacional_pib_total, color='r', linestyle='--', label=f'Média Nacional: {media_nacional_pib_total:.2f}')
plt.title('PIB Total por Região')
plt.legend()
plt.show()

# Calcular a média nacional do PIB per capita
media_nacional_pib_per_capita = df['PIB per capita'].mean()

# Gráfico para PIB per capita por região
sns.barplot(x='Região', y='PIB per capita', data=df, estimator=np.mean, ci=None)
plt.axhline(media_nacional_pib_per_capita, color='r', linestyle='--', label=f'Média Nacional: {media_nacional_pib_per_capita:.2f}')
plt.title('PIB per Capita por Região')
plt.legend()
plt.show()
```

```python
# Criar um dicionário com os novos nomes das colunas
new_column_names = {
    'Cód.': 'Codigo',
    'Município': 'Municipio',
    'PIB (Mil Reais)': 'PIB',
    'Impostos (Mil Reais)': 'Impostos',
    'VA da agropecuária (Mil Reais)': 'VA_Agropecuaria',
    'VA da indústria (Mil Reais)': 'VA_Industria',
    'VA dos serviços (Mil Reais)': 'VA_Servicos',
    'VA da administração pública (Mil Reais)': 'VA_AdminPublica'}

df_pib = df_pib.rename(columns=new_column_names)
df_pib['Região'] = df['Região']
df_pib['Nome_UF'] = df['Nome_UF']
df_pib.head()
```

```python

# Agrupar os dados por região e somar os valores de cada setor:
grouped_df = df_pib.groupby('Região').sum()

labels = grouped_df.index
va_agro_values = grouped_df['VA_Agropecuaria']
va_ind_values = grouped_df['VA_Industria']
va_serv_values = grouped_df['VA_Servicos']
va_adm_values = grouped_df['VA_AdminPublica']
tax_values = grouped_df['Impostos']

fig, ax = plt.subplots(figsize=(12, 8))

colors = sns.color_palette("pastel")

ax.bar(labels, va_agro_values, color=colors[0], label='VA Agropecuária')
ax.bar(labels, va_ind_values, bottom=va_agro_values, color=colors[1], label='VA Indústria')
ax.bar(labels, va_serv_values, bottom=va_agro_values + va_ind_values, color=colors[2], label='VA Serviços')
ax.bar(labels, va_adm_values, bottom=va_agro_values + va_ind_values + va_serv_values, color=colors[3], label='VA Administração Pública')
ax.bar(labels, tax_values, bottom=va_agro_values + va_ind_values + va_serv_values + va_adm_values, color=colors[4], label='Impostos')

ax.set_xlabel('Região')
ax.set_ylabel('Valor (Mil Reais)')
ax.set_title('Contribuição do VA e dos Impostos para o PIB por Região')
ax.legend()

fig.tight_layout()
plt.xticks(rotation=90)

plt.show()
```

# 6. Conclusão

Podemos destacar que apesar do PIB Brasileiro ser um dos maiores do planeta, ele é extremamente concentrada, podendo ser dividido em dois graus de concentração, dentro dos estados mais pobres, a capital concentra boa parte do PIB em comparação com o restante dos seus territórios. Enquanto isso o Sudeste concentra a grande maior parte do bolo do produto interno, em especial São Paulo, podemos perceber que a concentração dentro dos estados não é tão absurda como nos estados mais pobres, existem diversos polos onde o pib é bastante elevado e distribuído.

Podemos perceber como a população está diretamente correlacionada com o pib elevado de algumas regiões, visto que o setor de serviços é proeminente. E por conta disso regiões com mais industrias e menos populações tendem a possuir o PIB per capita mais elevado, o que não quer dizer exatamente que a região é menos desigual.

Por fim, como sem surpresas, SIM! o Brasil é um país desigual, e muito desigual, em diversos graus. Temos diversos cenários dentro do País, que demonstram como somos heterogêneos quando o assunto é qualidade de vida e recursos econômicos e financeiros.
